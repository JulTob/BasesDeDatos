## Ejercicio c)
Obtener una lista donde se indique cuantos empleados tiene comisi√≥n y cuantos no.

> Nota: El planteamiento tiene cierta ambiguedad. Por un lado en entorno t√©cnico, entender√≠amos que son los valores nulos en comisiones, _peeeero_ en entornos laborales, como en ofim√°tica, esta pregunta solicitar√≠a tambi√©n aquellos que tienen un valor de 0 en el campo no-nulo. Es interesante investigar ambos planteamientos. Voy a tomar esta ruta dual.

1. E := ùúå[EMPLOYEE]
2. A := ùúé(COMISSION is ‚êÄ)[E]
2. B := ùúé(COMISSION is 0)[E]
2. C := ùúé(COMISSION > 0)[E]
3. ‚èé R := |A| ‚à™ |B| ‚à™ |C|

````python
SQL("""
SELECT
    COUNT(*) AS TOTAL,
    COUNT(COMMISSION_PCT) AS CON_COMISION,
    COUNT(*) - COUNT(COMMISSION_PCT) AS SIN_COMISION
FROM EMPLOYEES;
""")
````

Truco: Se puede hacer un **histograma** anal√≠tico de los datos con la funci√≥n de COUNT aplicada a un GROUP-BY. Esto separa los tipos de la columna, con cada count aplicada a cada grupo en lugar de al total de tabla.

````python
SQL("""
SELECT
    COMMISSION_PCT as Comision,
    COUNT(COMMISSION_PCT) as Conteo
FROM EMPLOYEES
GROUP BY COMMISSION_PCT
;
""")
````

Esta soluci√≥n la marco como _Truco_ porque en otros contextos o en valores continuos de Commission_PCT podr√≠a potencialmente crear una tabla enorme. Aunque tambi√©n podr√≠amos extraer la informaci√≥n de esa tabla con `where`.

Alternativamente, la forma m√°s meticulosa ser√≠a as√≠:

````python
SQL("""
SELECT
    COUNT(COMMISSION_PCT) as "Known",
    SUM(COMMISSION_PCT IS 0) as "Comission=0",
    SUM(COMMISSION_PCT IS NOT NULL
        AND COMMISSION_PCT IS NOT 0) as "Comission>0",
    SUM(COMMISSION_PCT IS NULL) as "Unknown",
    SUM(COMMISSION_PCT) as "Total Comissions"

FROM EMPLOYEES
;
""")
````

> Tenemos que usar SUM(condici√≥n), donde cada entrada evalua la condici√≥n en esa instancia. En caso afirmativo devuelve 1, con lo que se suma al tallaje total.

> PD: Sabemos que solo podemos tener comisiones >0 por l√≥gica del negocio (Bussiness Logic).

> Si vemos el problema por partes, se puede resolver utilizando tablas intermedias. Una de ellas deber√≠a de identificar los Subalternos de cada Supervisor. La otra, los directores Supervisados.

````python
SQL("""
    /* Subordinados NO directores */
    SELECT
        M.employee_id AS Id,
        COUNT(*)      AS NumSub,
        0             AS NumDir
    FROM
        EMPLOYEES AS M,
        EMPLOYEES AS E
    WHERE
        M.employee_id = E.manager_id
        AND E.employee_id NOT IN (
            SELECT Dep.manager_id
            FROM DEPARTMENTS AS Dep
            WHERE Dep.manager_id IS NOT NULL
        )
    GROUP BY
        M.employee_id

    UNION ALL

    /* Subordinados que S√ç son directores */
    SELECT
        M.employee_id AS Id,
        0             AS NumSub,
        COUNT(*)      AS NumDir
    FROM
        EMPLOYEES AS M,
        EMPLOYEES AS E
    WHERE
        M.employee_id = E.manager_id
        AND E.employee_id IN (
            SELECT Dep.manager_id
            FROM DEPARTMENTS AS Dep
            WHERE Dep.manager_id IS NOT NULL
        )
    GROUP BY
        M.employee_id
""")
````

````python
SQL("""
SELECT
    Super.last_name  AS "Apellido Supervisor",
    Super.first_name AS "Nombre Supervisor",
    SUM(Supervisados.Sub) AS "N¬∫ Subordinados",
    SUM(Supervisados.Dir) AS "N¬∫ Directores Supervisados"
FROM
    EMPLOYEES AS Super,
    (
        SELECT
            M.employee_id AS Id,
            COUNT(*)      AS Sub,
            0             AS Dir
        FROM EMPLOYEES AS M, EMPLOYEES AS E
        WHERE
            M.employee_id = E.manager_id
            AND E.employee_id NOT IN (
                SELECT Dep.manager_id
                FROM DEPARTMENTS AS Dep
                WHERE Dep.manager_id IS NOT NULL
            )
        GROUP BY M.employee_id

        UNION ALL

        SELECT
            M.employee_id AS Id,
            0             AS Sub,
            COUNT(*)      AS Dir
        FROM EMPLOYEES AS M, EMPLOYEES AS E
        WHERE
            M.employee_id = E.manager_id
            AND E.employee_id IN (
                SELECT Dep.manager_id
                FROM DEPARTMENTS AS Dep
                WHERE Dep.manager_id IS NOT NULL
            )
        GROUP BY M.employee_id
    ) AS Supervisados
WHERE
    Super.employee_id = Supervisados.Id
GROUP BY
    Super.employee_id, Super.last_name, Super.first_name
ORDER BY
    Super.last_name, Super.first_name
;


""")
````
